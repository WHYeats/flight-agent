```python
import csv

AIRPORTS_FILE = "airports_iata.csv"

with open(AIRPORTS_FILE, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

small_airports = [row for row in rows if row["type"] == "small_airport"]

print(f"Total airports in airports_iata.csv : {len(rows)}")
print(f"Small airports                      : {len(small_airports)}")
```
